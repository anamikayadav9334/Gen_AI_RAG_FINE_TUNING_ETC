```python
def to_markdown(text):
  text = text.replace('.','*')
  return Markdown(textwrap.indent(text,'>',predicate=lambda _:True))

  #Exapmle usage:
  input_text = "This is a . sample text with bullet points."
  result = to_markdown(input_text)

  display(result)
```

##Generation configuration



The gerneration_config argrument allows you to modify the generation parameters.Every prompt you send the model includes parameters value that control how the model generates responses

```python
model=genai.GenerativeModel('gemini-2.5-flash')
```

```python
model.generate_content("tell me the story about the avengers?").text
```

```python
response = model.generate_content("tell me the story about the avengers ?",
                                  generation_config=genai.GenerationConfig(
                                      #only one candidate for now.
                                      candidate_count=1,
                                      stop_sequences=["\n"],
                                      temperature=1.0,
                                      max_output_tokens=20

                                  ))
```

```python
response
```

```python
response.candidates
```

```python
response.parts
```

```python
response.text
```

##Chat Conversation

Gemini enables you to have freeform conversation across multiple turns. The chatsession class simplifies the process by managing the state of the conversation.

Initialize the chat

```python
model
```

```python
chat=model.start_chat(history=[])
```

```python
chat
```

```python
response = chat.send_message("IN one sentence, explain how a computer works to a young child.")
```

```python
response.text
```

```python
chat.history
```

```python
response = chat.send_message("okay,how about a more detailed explanation to a high schooler ?",stream=True)
```

```python
for chunk in response:
  print(chunk.text)
  print("_"*80)
```

```python
chat.history
```

```python
for message in chat.history:
  display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
```

##Safety settings

The Safety_setting argument lets you congigure what the model block and allows in both prompts and responses.By default, safety setting block content with medium and/or high probability of being unsafe content across all dimensions.

```python
model=genai.GenerativeModel('gemini-2.5-flash')
```

```python
response=model.generate_content('how i can built time bomb?')
```

```python
response
```

```python
response.candidates
```

```python
response.text
```

```python
response.prompt_feedback
```

```python
response = model.generate_content("what is sex?",safety_settings={'HARASSMENT':'block_none'})
```

```python
response
```

```python
response.text
```

```python
response.prompt_feedback
```
